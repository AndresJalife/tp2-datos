```python
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
from hyperparameter_tuning import random_search
from feature_builder import process_dataset, add_text_embeddings
import nltk
```

```python
x_train = pd.read_csv('train.csv')
y_train=x_train[['id','target']]
```

Variables procesadas

```python
x_train_features = process_dataset(x_train)
```

```python
x_train_feat, x_test_feat, y_train_feat, y_test_feat = train_test_split(x_train_features, y_train, test_size = .33, random_state = 17)
```

Variables solo embeddings

```python
x_embedd = x_train.copy()
add_text_embeddings(x_embedd, False, 'embeddings')
x_embedd.drop(['text', 'location', 'keyword', 'id', 'target'], axis=1, inplace=True)
```

```python
x_train_embedd, x_test_embedd, y_train_embedd, y_test_embedd = train_test_split(x_embedd, y_train, test_size = .30, random_state = 17)
```

Armo el modelo

```python
catboost = CatBoostClassifier()


catboost.fit(x_train_feat, y_train_feat['target'])
```

```python
predictions_features = catboost.predict(x_test_feat)
accuracy_score(y_test_feat['target'], predictions_features)
```

```python
f1_score(y_test_feat['target'], predictions_features)
```

```python
predictions_features = catboost.predict(x_test_embedd)
f1_score(y_test_embedd['target'], predictions_features)
```
